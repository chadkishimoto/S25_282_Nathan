**Exercise:** Create a heat map plot of the function
$$ f(x, y) = e^{- (x^2 + y^2)}$$
Using:
- 50 points in x from -2 to +2
- 200 points in y from -4 to +4

```python
x = np.linspace(-2,2,50)
y = np.linspace(-4,4,200)
N = np.zeros((len(y), len(x)))
def g(x,y): 
    return np.exp(-(x**2+y**2))
print(len(x),len(y))
for i in range(len(y)): 
    for j in range(len(x)): 
        N[i,j] = g(x[j],y[i])
plt.figure(figsize=(8,8))
plt.imshow(N,extent=[-2,2,-4,4]) #<== Replace MatrixName with your matrix's name
plt.colorbar()
plt.show()
```

**Exercise:** Create a contour plot of $g(x,y)$ with the same x and y arrays.
- Your contours should be at 75%, 50%, 25% of the maximum value
- Your contours should be labeled with their values
- You should include a symbol at the location of the maximum
- Your code should be completely self-contained, which means you haven't "hard coded" in any values.

```python
def w(x,y): 
    return np.exp(-x**2)*np.exp(-(y**2)/4)
x = np.linspace(-2,2,50)
y = np.linspace(-4,4,200)
N = np.zeros((len(y), len(x)))
for i in range(len(y)): 
    for j in range(len(x)): 
        N[i,j] = w(x[j],y[i])
maximum_index = np.unravel_index(np.argmax(N), N.shape)
plt.figure(figsize=(8,8))
CS = plt.contour(x,y, N, levels = [np.max(N)*0.25, np.max(N)*0.50, np.max(N)*0.75]) 
plt.plot(x[maximum_index[1]], y[maximum_index[0]], '*')
plt.colorbar()
plt.clabel(CS)
plt.show()
```
