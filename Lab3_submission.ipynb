**Exercise:** Use the data from ``Lab3Data.csv`` to create the following plots:
- a plot of $\chi^2$ vs $m$
- a plot that shows the data points, with error bars, as well as the best fit model and the upper and lower bounds models

In the cell below, place the code that creates the $\chi^2$ vs $m$ plot

```python
indep, data, uncertainties = np.loadtxt('Lab3Data.csv',delimiter=',',usecols=(0,1,2), unpack=True)
def chi2(m):
    y=sum(((data-m*indep)**2)/(uncertainties**2))
    return y
        
m_values=np.linspace(3,7,100)
chi_values = np.zeros(len(m_values))
for i in range(len(m_values)):
    chi_values[i] = chi2(m_values[i])
plt.figure()
plt.plot(m_values,chi_values)
plt.xlabel('m value')
plt.ylabel('Chi Squared')
plt.show()
```

In the cell below, place the code (that utilizes code from above) to create the plots that include: data points with error bars; model curves for the best fit and the upper and lower bound slopes.

```python
x = np.linspace(np.min(indep),np.max(indep))
def f(x,m): 
    return m*x 
y = f(x,5.3)
w = f(x,5.0)
z = f(x, 5.6)
indep, data, U = np.loadtxt('Lab3Data.csv',delimiter=',',usecols=(0,1,2), unpack=True)
plt.figure()
plt.plot(x,y, linestyle = '-', label = 'm=5.3')
plt.plot(x,w, linestyle = '--', label = 'm=5.0')
plt.plot(x,z, linestyle = '-.', label= 'm=5.6')
plt.legend(loc='upper right')
plt.xlabel('indep variables')
plt.ylabel('data')
plt.xlim(0,1)
plt.errorbar(indep, data, yerr = uncertainties, fmt='o')
plt.show()
```

Use a ``print`` statement to print the value of the best fit $m$ and its uncertainty

```python
print('The best fit for m is 5.3 with an uncertainty of 0.3')
```

In the cell below, place the code that creates the $\chi^2$ vs $\lambda$ plot

```python
n_values, y_values, Uncertainty = np.loadtxt('Lab3bData.csv',delimiter=',',usecols=(0,1,2), unpack=True)
n_values, y_values, Uncertainty = np.loadtxt('Lab3bData.csv',delimiter=',',usecols=(0,1,2), unpack=True)
d = 0.125 / 1000 
L = 280 / 100 
def M(n, lambda_in_nm):
    lambda_m = lambda_in_nm * 1e-9  
    theta = np.arcsin(np.clip((n * lambda_m) / d, -1, 1))  
    return L * np.tan(theta) * 1000  
def CHI2(lambda_in_nm):
    chi2 = np.sum(((y_values - M(n_values, lambda_in_nm)) ** 2) / (Uncertainty ** 2))
    return chi2
lambda_range = np.linspace(400, 700, 1000)
chi2_values = np.array([CHI2(lmbd) for lmbd in lambda_range])
plt.figure()
plt.plot(lambda_range, chi2_values)
plt.xlabel('Wavelength (nm)')
plt.ylabel(r'$\chi^2$ Value')
plt.title('Chi-Squared vs Wavelength')
```
