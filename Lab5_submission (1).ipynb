```python
indep, data, uncertainties = np.loadtxt('Lab3Data.csv',delimiter=',',usecols=(0,1,2), unpack=True)
def chi2(m,b):
    y=sum(((data-(m*indep+b))**2)/(uncertainties**2))
    return y


m=np.linspace(4,7,100)
b=np.linspace(-0.5,0.5,200)
N = np.zeros((len(b), len(m)))
print(len(m),len(b))
for i in range(len(b)): 
    for j in range(len(m)): 
        N[i,j] = chi2(m[j],b[i])



print(np.unravel_index(np.argmin(N), N.shape))
print('m =',m[44], 'b =',b[93])
plt.figure(figsize=(8,8))
CS = plt.contour(m,b, N, levels = [np.min(N), np.min(N)*2.3, np.min(N)*6.17]) 
plt.colorbar()
plt.clabel(CS)
plt.show()


indep, data, uncertainties = np.loadtxt('Lab3Data.csv',delimiter=',',usecols=(0,1,2), unpack=True)
F= np.zeros(len(indep))
Fit_curve = (m[44]*indep + b[93]) 

plt.figure()
plt.plot(indep, Fit_curve)
plt.xlabel('x')
plt.ylabel('y')
plt.errorbar(indep, data, yerr = uncertainties, fmt='o')
plt.show()
```

**Exercise:**  Using ``interference.py``, create a heat map from double slit interference

```python
import interference as Cint
Y=np.linspace(-50,50,200)
L=np.linspace(0,2,200)
K = np.zeros((len(Y),len(L)))

for i in range(len(Y)): 
    for j in range(len(L)): 
        K[i,j] = Cint.DoubleSlitIntensity(Y[i],0.05,L[j],400)
plt.figure(figsize=(8,8))
plt.imshow(K,extent=[0,2,-50,50], aspect=0.02)
plt.colorbar()
plt.show()
```
