Create a function that takes an integer, ``N``, as an input and outputs an array of length N that represents a random walk using either +1 or -1 as the step.

```python
def random_walk(N):
    location = np.zeros(4)
    loc = np.zeros(N)
    location[0] = 0 
    for i in range(N):
        loc[i] = loc[i-1] + rng.choice((-1,1))
    return loc
plt.figure()
plt.plot(random_walk(100), '-o')
plt.show()
    
```
